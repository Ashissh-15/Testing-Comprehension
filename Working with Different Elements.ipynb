```python
#pressing an icon in burger king website

from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver = webdriver.Edge()
driver.get("https://www.burgerking.in/")

try:
    hamburger_icon = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "img[alt='HamburgerLogo']")))
    
    hamburger_icon.click()

    time.sleep(10)

except Exception as e:
    print("Error:", e)

finally:
    driver.quit()
```

```python
# Handling alerts
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

driver = webdriver.Edge()
driver.get("https://www.dezlearn.com/javascript-alerts/")
time.sleep(10)

driver.find_element(By.ID,"s_alert1").click()
alert=driver.switch_to.alert
time.sleep(3)
alert.accept()
time.sleep(3)

# driver.switch_to.frame("ml-webforms-popup-3672985")
# driver.find_element(By.CLASS_NAME ,"close").click()
# driver.switch_to.parent_frame()

c2=driver.find_element(By.ID,"c_alert2")
driver.execute_script("arguments[0].scrollIntoView();", c2)
c2.click()
alert=driver.switch_to.alert
time.sleep(3)
alert.dismiss()
time.sleep(3)

driver.find_element(By.ID,"p_alert3").click()
alert=driver.switch_to.alert
time.sleep(3)
alert.send_keys("tokyo")
alert.accept()
time.sleep(3)

driver.quit()
```
