```python
#to search for something and then collect the data given along with taking a screenshot
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
driver = webdriver.Edge()
driver.get("https://burgerking.in")

try:
    hamburger_icon = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.CSS_SELECTOR, "img[alt='search']")))
    driver.save_screenshot("screenshot1_testing.png")
    hamburger_icon.click()
    
    search_field = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.ID, "searchValue")))
    search_field.send_keys("Veg Burger")
    
    try:
        data=driver.find_element(By.CLASS_NAME , "search-page")
        driver.implicitly_wait(15)
        driver.save_screenshot("screenshot2_testing.png")
    
        print(data.text)

        time.sleep(15)

    except Exception as e:
        print("Error:", e)

except Exception as e:
        print("Error:", e)

finally:
    driver.quit()


```

```python
#get one of the panel values from amazon
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

driver = webdriver.Edge()
driver.get("https://www.amazon.in/")

time.sleep(15)
element = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//a[@id='nav-hamburger-menu']")))

element.click()

mob = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "(//a[@class='hmenu-item'])[10]")))
mob.click()
val = driver.find_element(By.CLASS_NAME, "hmenu-translateX")
print(val.text)

time.sleep(10)


```
