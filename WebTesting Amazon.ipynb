```python
# Grid code to run multiple browsers at the same time
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import time

class ReadConfigurations:
    @staticmethod
    def read_configuration(browser):
        global drivers


        if browser == "firefox":
            desired_cap = webdriver.FirefoxOptions()
            drivers['firefox'] = webdriver.Remote(command_executor="http://172.20.10.2:4444", options=desired_cap)
        elif browser == "edge":
            desired_cap = webdriver.EdgeOptions()
            drivers['edge'] = webdriver.Remote(command_executor="http://172.20.10.2:4444", options=desired_cap)
        else:
            raise ValueError("Invalid browser name. Please choose from  'firefox', or 'edge'.")

try:
    drivers = {}
    browsers = ['firefox', 'edge']
    for browser in browsers:
        ReadConfigurations.read_configuration(browser)

    for browser, driver in drivers.items():
        driver.get("https://www.amazon.in/")
        driver.maximize_window()
        services_link = driver.find_element(By.LINK_TEXT, "Mobiles")
        services_link.click()
        
        shop = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.LINK_TEXT, "Cameras")))
        driver.execute_script("arguments[0].scrollIntoView();", shop)
        shop.click()

        driver.back()
        driver.back()

    time.sleep(20)

    for driver in drivers.values():
        driver.quit()

except ValueError as e:
    print(e)
```
