```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import csv

driver = webdriver.Firefox()
success=0
total=0

driver.get("https://vit.ac.in/detailview/bike-stunt-show")

with open("image_final_data.csv", mode="r") as file:
    reader = csv.reader(file)
    for row in reader:
        xpath = row[0]
        
        expected_image_url = row[1]

        image_element = driver.find_element(By.XPATH, xpath)

        displayed_image_url = image_element.get_attribute("src")

        if displayed_image_url == expected_image_url:
            print("The displayed image matches the expected image for corresponding XPath")
            success+=1
        else:
            print("The displayed image does not match the expected image for corresponding XPath")
        total+=1
print("Success Rate is ",success*100/total,"%")
# Close the browser
driver.quit()
```
